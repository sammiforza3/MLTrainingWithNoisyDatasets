```python
import pandas as pd

# Creazione DataFrame di esempio
df = pd.DataFrame({
    'città': ['Roma', 'Milano', 'Roma', 'Napoli', 'Milano'],
    'voto': ['A', 'B', 'A', 'C', 'B']
})

# Conversione singola colonna
df['città'] = df['città'].astype('category')

# Conversione di più colonne contemporaneamente
cols_to_fix = ['città', 'voto']
df[cols_to_fix] = df[cols_to_fix].astype('category')
```

```python
df2 = pd.DataFrame()
df2["città"]=df["città"]
df2["città codificata"] = df["città"].cat.codes
```
