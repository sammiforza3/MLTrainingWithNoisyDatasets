```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#file_path = "Datasets/Teen_Mental_Health_Dataset.csv"

import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("algozee/teenager-menthal-healy")

print("Path to dataset files:", path)

file_path = os.path.join(path, "Teen_Mental_Health_Dataset.csv")

df = pd.read_csv(file_path)

df.head()
```

```python
#trasformo la depression label in categorica

df["depression_label"] = df["depression_label"].astype("category")
```

```python
#plottiamo la classe depression_label

plt.pie(df["depression_label"].value_counts(), labels=[0,1])
plt.legend()
plt.title("distribuzione della variabile depression_label")
plt.show()
```
